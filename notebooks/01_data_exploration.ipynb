```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# Simulated dataset (multiple images)
urls = [
    "https://picsum.photos/300",
    "https://picsum.photos/301",
    "https://picsum.photos/302"
]

images = []

# Load images
for url in urls:
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))
    images.append(img)

# Show images
for i, img in enumerate(images):
    plt.figure()
    plt.imshow(img)
    plt.title(f"Image {i+1}")
    plt.axis('off')
```

## Loading Multiple Images

This step simulates loading multiple images from a dataset.

```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# Simulated dataset with labels
dataset = [
    {"url": "https://picsum.photos/300", "label": "animal"},
    {"url": "https://picsum.photos/301", "label": "animal"},
    {"url": "https://picsum.photos/302", "label": "animal"}
]

images = []
labels = []

# Load images and labels
for item in dataset:
    response = requests.get(item["url"])
    img = Image.open(BytesIO(response.content))
    images.append(img)
    labels.append(item["label"])

# Display with labels
for i, img in enumerate(images):
    plt.figure()
    plt.imshow(img)
    plt.title(f"Label: {labels[i]}")
    plt.axis('off')
```
